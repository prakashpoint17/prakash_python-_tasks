```python
numbers = [10, 20, 30, 40]
print(numbers[0])      # 10
numbers.append(50)     # Add element
numbers[1] = 25        # Modify element
print(numbers)
```

```python
numbers.append(60)
numbers.insert(1, 15)
numbers.remove(30)
numbers.pop()
numbers.sort()
numbers.reverse()
```

```python
nums = {1, 2, 3, 3, 4}
print(nums)   # {1, 2, 3, 4}
```

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a | b)   # Union
print(a & b)   # Intersection
print(a - b)   # Difference
```

```python
try:
    num = int(input("Enter number: "))
    print(10 / num)

except ZeroDivisionError:
    print("Cannot divide by zero!")

except ValueError:
    print("Invalid input!")

finally:
    print("Done")
```
